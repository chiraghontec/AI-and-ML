```python
import pandas as pd
import seaborn as sns
import matplotlib 
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
```

```python
#Boxplot 
plt.title('box plot')
sns.boxplot(df['Price'])
plt.show()
```

```python
#Histogram to represent the milage of the car 
plt.title('Miles')
plt.hist(df['MPG.city'])
plt.show()
```

```python
#scatter plot on the horse power of the vehicles
plt.title('Scatter plot')
plt.scatter(df['Horsepower'], df['MPG.city'])
plt.xlabel('Horse Power')
plt.ylabel('Milage')

plt.show()
```

```python
#line chart of the cylinder and miles per gallon of the vehicles
plt.title('Line chart')
sns.lineplot(x="EngineSize", y="MPG.city", data=df)
```
